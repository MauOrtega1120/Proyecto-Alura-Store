### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt

# Creamos una lista con los nombres de las tiendas para facilitar el trabajo
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Calculamos la suma total de la columna 'Precio' para cada tienda
# Usamos las variables tienda, tienda2, tienda3 y tienda4 que ya están cargadas
facturacion_por_tienda = [
    tienda['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]

# Creamos un gráfico de barras para visualizar los resultados
plt.figure(figsize=(10, 5))
plt.bar(nombres_tiendas, facturacion_por_tienda, color=['#3498db', '#9b59b6', '#e67e22', '#e74c3c'])
plt.title('Facturación Total por Tienda')
plt.ylabel('Monto Total ($)')
plt.show()

# Mostramos los valores exactos
for i in range(len(nombres_tiendas)):
    print(f"{nombres_tiendas[i]}: ${facturacion_por_tienda[i]:,.2f}")
```

# 2. Ventas por categoría

```python
# Usamos .value_counts() para contar cuántos productos hay por cada categoría
# Mostramos el Top 3 de cada tienda para comparar
print("--- Categorías más populares ---")

print("\nTienda 1:")
print(tienda['Categoría del Producto'].value_counts().head(3))

print("\nTienda 2:")
print(tienda2['Categoría del Producto'].value_counts().head(3))

print("\nTienda 3:")
print(tienda3['Categoría del Producto'].value_counts().head(3))

print("\nTienda 4:")
print(tienda4['Categoría del Producto'].value_counts().head(3))
```

# 3. Calificación promedio de la tienda


```python
# Calculamos el promedio simple con .mean()
promedios_calificacion = [
    tienda['Calificación'].mean(),
    tienda2['Calificación'].mean(),
    tienda3['Calificación'].mean(),
    tienda4['Calificación'].mean()
]

# Graficamos para ver la diferencia visualmente
plt.figure(figsize=(10, 5))
plt.bar(nombres_tiendas, promedios_calificacion, color='gold')
plt.title('Calificación Promedio de los Clientes')
plt.ylim(0, 5) # Ponemos el límite de 0 a 5 estrellas
plt.ylabel('Puntaje Promedio')
plt.show()

for i in range(len(nombres_tiendas)):
    print(f"{nombres_tiendas[i]}: {promedios_calificacion[i]:.2f} estrellas")
```

# 4. Productos más y menos vendidos

```python
# Definimos una función pequeña para no repetir el mismo código 4 veces
def mostrar_productos(df, nombre):
    print(f"--- Productos en {nombre} ---")
    print("Top 3 Más Vendidos:")
    print(df['Producto'].value_counts().head(3))
    print("3 Menos Vendidos:")
    print(df['Producto'].value_counts().tail(3))
    print("-" * 30)

mostrar_productos(tienda, "Tienda 1")
mostrar_productos(tienda2, "Tienda 2")
mostrar_productos(tienda3, "Tienda 3")
mostrar_productos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Calculamos el promedio del costo de envío
costos_envio = [
    tienda['Costo de envío'].mean(),
    tienda2['Costo de envío'].mean(),
    tienda3['Costo de envío'].mean(),
    tienda4['Costo de envío'].mean()
]

plt.figure(figsize=(10, 5))
plt.bar(nombres_tiendas, costos_envio, color='lightgreen')
plt.title('Costo Promedio de Envío')
plt.ylabel('Costo en Pesos ($)')
plt.show()

for i in range(len(nombres_tiendas)):
    print(f"{nombres_tiendas[i]}: ${costos_envio[i]:.2f}")
```
